对于一个M × M 的方阵A，特征向量方程的定义为 $Au_i = λ_iu_i \tag{C.29}$ 其中i = 1, . . . , M，ui被称为特征向量(eigenvector)，λi被称为对应的特征值(eigenvalue)。 
这可以看成M 个齐次线性方程组，解存在的条件为
$|A − λiI| = 0 \tag{C.30}$ 
这被称为特征方程(characteristic equation)。**由于这是$λ_i$的M阶多项式，因此它一定有M个解 (虽然这些解未必不同)**。A的秩等于非零特征值的个数。（前提是A是可对角化矩阵）

### 补充
- 可对角化矩阵：wiki定义如果一个方块矩阵 A 相似于对角矩阵，也就是说，如果存在一个可逆矩阵 P 使得 $P^{−1}AP$ 是对角矩阵，则它就被称为可对角化的。这个矩阵的对角元素是 A 的特征值。
- 相似关系是两个矩阵之间的一种等价关系。两个n×n矩阵A与B为相似矩阵当且仅当存在一个n×n的可逆矩阵P，使得：$P^{−1}AP=B$, P被称为矩阵A与B之间的相似变换矩阵。
- 矩阵和可逆矩阵相乘，秩不变。
- 可逆矩阵可以用初等变换矩阵给表示出来（求逆的过程，就是增广矩阵+初等变换）。初等变换矩阵不改变秩。

我们特别感兴趣的是对称矩阵。协方差矩阵、核矩阵、Hessian矩阵都是对称矩阵。$对阵矩 阵的性质为A_{ij} = A_{ji}或者等价地，A = A^T 。对称矩阵的逆矩阵也是对称的。将A^T A = I取转 置，然后使用AA^{−1} = I$以及I的对称性即可证明这一点。 


通常情况下，矩阵的特征值是复数。但是对于对称矩阵，$特征值λ_i 为实数。这点可以用下面 的方式证明。首先将公式(C.29)左乘(u^∗_i )^T$ ，其中∗表示复共轭，我们可以得到 

$(u_i)^T Au_i = λ_i(u^∗_i )^T u_i \tag{C.31} $

之后，我们对公式(C.29)取复共轭，然后左乘$u^T_i$ ，可得 

$u_i^T Au_i^* = λ_i^*u_i ^T u_i^* \tag{C.32} $

推导过程中，我们使用了$A^∗ = A$，因为我们只考虑实对称矩阵A。将第二个方程取转置，使用$A^T = A$，我们看到两个方程的左侧相同，从而$λ^∗_i = λ_i$，因此$λ_i$一定是实数。 （2.3高斯分布，习题2.18有推导）

```python
# 推导实数
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 10))
d = plt.imread('./graph/2.3推导实数.png')
plt.axis('off')
ax.imshow(d)
```

实对称矩阵的特征向量$u_i$ 可以被选成单位正交的(即正交的并且长度为单位长度)，使得 
$u^T_i u_j = I_{ij} \tag{C.33}$
其中Iij是单位矩阵I的元素。为了证明这一点，我们首先将公式(C.29)左乘$u^T_j$ ，得到 

$u^T_j Au_i = λ_iu^T_j u_i \tag{C.34}$
因此，通过交换下标，我们有 

$u^T_i Au_j = λ_ju^T_i u_j \tag{C.35}$

我们现在对第二个方程取转置，使用对称性质$A^T = A$，然后将两个方程相减，可得 $( λ_i − λ_j ) u^T_i u_j = 0$

因此，
- 对于$ λ_i \neq λ_j$ ，我们有$u^T_i u_j = 0$，因此u^i和u^j 是正交的。
- 如果两个特征值是相等的，那 么任意线性组合$αu_i + βu_j$也是一个有着相同特征值的特征向量，因此我们可以任意选择一个线性组合，然后选择第二个特征向量正交于第一个(可以证明这种退化的特征向量永远不会线性相关)。(选正交的)

**因此特征向量可以选择为正交的，然后归一化为单位长度。**

**由于有M 个特征值，对应 的M 个特征向量组成了一个完备集，因此任意一个M 维的向量都可以表示为特征向量的线性组合**。 

```python
# 推导正交
fig, ax = plt.subplots(figsize=(10, 10))
d = plt.imread('./graph/2.3推导正交.png')
plt.axis('off')
ax.imshow(d)
```

我们可以令特征向量$u_i$组成成M × M的矩阵U，根据单位正交性，我们有
$U^TU=I \tag{C.37} $

这 样 的 矩 阵 被 称 为 正 交 的 (orthogonal)。 有 趣 的 是， 矩 阵 的 行 也 是 正 交 的， 即$UU^T = I$。为了证明这一点，我们注意到，公式(C.37)表明$U^T UU^{−1} = U^{−1} = U^T ，因 此UU^{−1} = UU^T = I$。使用公式(C.12)，也可以看出$|U| = 1$。 

特征向量方程(C.29)可以使用U 表示成下面的形式 
$AU = UΛ \tag{C.38}$

$其中Λ是一个M × M的对角矩阵，对角线上的元素为特征值λ_i$。

如果我们考虑一个列向量x，它经过正交矩阵U 进行变换，得到新向量 
$\tilde x = U x \tag{C.39}$
变换前后向量的长度不变，因为 
$$
\tilde{x}^T \tilde{x}=\boldsymbol{x}^{T} \boldsymbol{U}^{T} \boldsymbol{U} \boldsymbol{x}=\boldsymbol{x}^{T} \boldsymbol{x}  \tag{C.40}
$$

类似地，任意两个向量的角度在变换前后也不变，因为 ($cos\theta$不变)
$$
\tilde{x}^T \tilde{y}=\boldsymbol{x}^{T} \boldsymbol{U}^{T} \boldsymbol{U} \boldsymbol{y}=\boldsymbol{x}^{T} \boldsymbol{y}  \tag{C.41}
$$

因此，**乘以U 可以表示为坐标系的刚性旋转**。

根据公式(C.38)可得 
$ U^T AU = Λ\tag{C.42} $

因为Λ是对角矩阵，我们说矩阵A被矩阵U对角化(diagonalised)。如果我们左乘U然后右 乘$U^T$ ，我们有

$A = UΛU^T \tag{C.43} $

取这个方程的逆，然后使用公式(C.3)以及$U^{−1} =U^T$，我们有 
$ A^{-1}= UΛ^{-1}U^T  \tag{C.44}$
最后两个方程也可以写成 
􏰒
$$ A = \sum\limits_{i=1}^M\lambda_iu_iu_i^T \tag{C.45} $$

$$ A^{-1} = \sum\limits_{i=1}^M\frac{1}{\lambda_i}u_iu_i^T \tag{C.46} $$

协方差矩阵$ \Sigma $可以表示成特征向量的展开的形式：

$$ \Sigma = \sum\limits_{i=1}^D\lambda_iu_iu_i^T \tag{2.48} $$

同样的协方差的逆矩阵$ \Sigma^{-1} $可以表示为

$$ \Sigma^{-1} = \sum\limits_{i=1}^D\frac{1}{\lambda_i}u_iu_i^T \tag{2.49} $$

```python
# 推导展开
fig, ax = plt.subplots(figsize=(10, 10))
d = plt.imread('./graph/2.3推导展开.png')
plt.axis('off')
ax.imshow(d)
```

如果我们取公式(C.43)的行列式，然后使用公式(C.12):|AB|=|A||B|，我们有 (因为U是单位矩阵)
$$
|\boldsymbol{A}|=\prod_{i=1}^{M} \lambda_{i} \tag{C.47}
$$
类似地，取公式(C.43)的迹，使用迹运算的循环性(C.8):Tr(AB)=Tr(BA)以及$U^T U = I$，我们有 
􏰒$$
\operatorname{Tr}(\boldsymbol{A})=\sum_{i=1}^{M} \lambda_{i} \tag{C.48}
$$

使用结论(C.33)、(C.45)、(C.46)和(C.47)，可以证明公式(C.22)，我们把证明留给 读者作为练习。 (在3.5节，3.117，习题3.21的时候讲了)

一个矩阵A被称为正定的(positive definite)，记作A ≻ 0，如果对于所有非零向量w都 有$w^T Aw$ > 0。等价地，一个正定矩阵的所有特征值都有$λ_i > 0$。令w为每一个特征向量，然后 注意到任意的向量都可以展开为特征向量的组合，我们即可以证明这一点。

注意，正定不同于 所有元素都为正。例如，矩阵 
􏰊$$
\left(\begin{array}{ll}
1 & 2 \\
3 & 4
\end{array}\right) \tag{C.49}
$$

的特征值为$λ_1 ≃ 5.37且λ_2 ≃ −0.37$。

一个矩阵被称为半正定的(positive semidefinite)，如果对 
于w的所有值都有$w_TAw ≥ 0，记作A ≽ 0。它等价于λ_i ≥ 0$。
